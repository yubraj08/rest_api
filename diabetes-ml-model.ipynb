```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
dia_data = pd.read_csv("/kaggle/input/pima-indians-diabetes-database/diabetes.csv")
```

```python
dia_data.head()
```

**Number of rows and columns**

```python
dia_data.shape
```

**Getting statistical measures of the data**

```python
dia_data.describe()
```

```python
dia_data['Outcome'].value_counts()
```

**0 -> non diabetic and 1 -> diabetic**

```python
dia_data.groupby('Outcome').mean()
```

```python
X = dia_data.drop(columns = 'Outcome',axis=1)
Y = dia_data['Outcome']
```

**axis = 1 means dropping column
axis = 2 means dropping row**

**Data Standardization**

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standard_data = scaler.transform(X)
```

```python
print(standard_data)
```

```python
X = standard_data
Y = dia_data['Outcome']
```

```python
print(X)
print(Y)
```

**Train Test Split**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2,stratify=Y,random_state=2)
```

train test split data into four variable i.e X and Y train and test,

test size is 0.2 for 20% data for testing,

stratify divides outcomes in various cases,

random state dividing data in different style

```python
print(X.shape,X_train.shape,X_test.shape)
```

**Training the model**

```python
classify = svm.SVC(kernel='linear')
```

**Training the support vector machine classifier**

```python
classify.fit(X_train,Y_train)
```

**Model Evaluation**

**Accuracy Score**

```python
X_train_predict = classify.predict(X_train)
train_acc = accuracy_score(X_train_predict,Y_train)
```

```python
print(train_acc)
```

```python
X_test_predict = classify.predict(X_test)
test_acc = accuracy_score(X_test_predict,Y_test)
```

```python
print(test_acc)
```

**Making a Predictive System**

```python
input_data = (3.36,31.97,19.35,15.95,15.24,7.88,0.33,11.76)


#changing to numpy array
input_data_as_numpy_array = np.asarray(input_data)


#reshape the array as this particular input data

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)


# input standardization

std_data = scaler.transform(input_data_reshaped)

print(std_data)



prediction = classify.predict(std_data)
print(prediction)

```
